```python
Descrição em Linguagem Algorítmica

Agora a mesma ideia, mas num formato mais “formalzinho”, daquele jeito que professor gosta:

ALGORITMO Gestao_Colaborativa
VAR
    opcao, subopcao: INTEIRO
    id_projeto: INTEIRO
    titulo, descricao, prazo, responsavel, status: CADEIA
    projetos: LISTA_DE_PROJETOS
    projeto: TIPO_PROJETO
    SAIR_DO_LACO: LOGICO

INÍCIO

    SAIR_DO_LACO ← FALSO

    ENQUANTO SAIR_DO_LACO = FALSO FAÇA

        LIMPAR_TELA()
        ESCREVA("===== GESTÃO COLABORATIVA =====")
        ESCREVA("Carregando lista de projetos...")

        projetos ← BUSCAR_TODOS_PROJETOS()

        SE LISTA_VAZIA(projetos) ENTÃO
            ESCREVA("Nenhum projeto cadastrado.")
        SENÃO
            ESCREVA("ID | TÍTULO | RESPONSÁVEL | STATUS")
            PARA CADA projeto EM projetos FAÇA
                ESCREVA(projeto.id, projeto.titulo, projeto.responsavel, projeto.status)
            FIMPARA
        FIMSE

        ESCREVA("")
        ESCREVA("Opções:")
        ESCREVA("1 - Criar novo projeto")
        ESCREVA("2 - Ver detalhes de um projeto")
        ESCREVA("3 - Editar um projeto")
        ESCREVA("4 - Excluir um projeto")
        ESCREVA("0 - Sair")
        LEIA(opcao)

        ESCOLHA opcao
            CASO 1:
                CHAMAR Cadastrar_Projeto()
            CASO 2:
                CHAMAR Detalhar_Projeto()
            CASO 3:
                CHAMAR Editar_Projeto()
            CASO 4:
                CHAMAR Excluir_Projeto()
            CASO 0:
                SAIR_DO_LACO ← VERDADEIRO
            CASO CONTRÁRIO:
                ESCREVA("Opção inválida.")
                PAUSAR()
        FIMESCOLHA

    FIMENQUANTO

    ESCREVA("Encerrando o sistema Gestão Colaborativa...")

FIM



```

```python
Procedimento 1 – Cadastrar_Projeto
PROCEDIMENTO Cadastrar_Projeto()
VAR
    titulo, descricao, prazo, responsavel, status: CADEIA
    projeto: TIPO_PROJETO
INÍCIO
    LIMPAR_TELA()
    ESCREVA("=== Novo Projeto ===")
    ESCREVA("Informe o título:")
    LEIA(titulo)
    ESCREVA("Informe a descrição:")
    LEIA(descricao)
    ESCREVA("Informe o prazo:")
    LEIA(prazo)
    ESCREVA("Informe o responsável:")
    LEIA(responsavel)
    ESCREVA("Informe o status:")
    LEIA(status)

    SE titulo = "" OU responsavel = "" ENTÃO
        ESCREVA("Erro: Título e responsável são obrigatórios.")
    SENÃO
        projeto ← CRIAR_PROJETO(titulo, descricao, prazo, responsavel, status)
        SALVAR_PROJETO(projeto)
        ESCREVA("Projeto criado com sucesso.")
    FIMSE

    PAUSAR()
FIMPROCEDIMENTO
```

```python
Procedimento 2 – Detalhar_Projeto
PROCEDIMENTO Detalhar_Projeto()
VAR
    id_projeto: INTEIRO
    projeto: TIPO_PROJETO
    tarefas: LISTA_DE_TAREFAS
    comentarios: LISTA_DE_COMENTARIOS
    subopcao: INTEIRO
INÍCIO
    LIMPAR_TELA()
    ESCREVA("Informe o ID do projeto:")
    LEIA(id_projeto)

    projeto ← BUSCAR_PROJETO_POR_ID(id_projeto)

    SE projeto = NULO ENTÃO
        ESCREVA("Projeto não encontrado.")
        PAUSAR()
    SENÃO
        tarefas ← BUSCAR_TAREFAS_DO_PROJETO(projeto)
        comentarios ← BUSCAR_COMENTARIOS_DO_PROJETO(projeto)

        ESCREVA("Título: ", projeto.titulo)
        ESCREVA("Descrição: ", projeto.descricao)
        ESCREVA("Prazo: ", projeto.prazo)
        ESCREVA("Responsável: ", projeto.responsavel)
        ESCREVA("Status: ", projeto.status)

        ESCREVA("Tarefas:")
        // (listagem das tarefas)

        ESCREVA("Comentários:")
        // (listagem dos comentários)

        ESCREVA("1 - Adicionar tarefa")
        ESCREVA("2 - Adicionar comentário")
        ESCREVA("Outra tecla - Voltar")
        LEIA(subopcao)

        SE subopcao = 1 ENTÃO
            CHAMAR Cadastrar_Tarefa(projeto)
        SENÃO SE subopcao = 2 ENTÃO
            CHAMAR Cadastrar_Comentario(projeto)
        FIMSE

    FIMSE
FIMPROCEDIMENTO


(Editar_Projeto e Excluir_Projeto seguem a mesma ideia do pseudocódigo.)
```
